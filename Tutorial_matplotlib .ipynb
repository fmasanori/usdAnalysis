# MATPLOTLIB.PYPLOT

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.DataFrame({
    'name':['john','mary','peter','jeff','bill','lisa','jose'],
    'age':[23,78,22,19,45,33,20],
    'gender':['M','F','M','M','M','F','M'],
    'state':['california','dc','california','dc','california','texas','texas'],
    'num_children':[2,0,0,3,2,1,4],
    'num_pets':[5,1,0,5,2,2,3]
})
```

```python
print([x for x in range(min(df.age.values), max(df.age.values)+1, 10 )])
```

```python
print(df)
```

### ploting scatter

```python
df.plot(kind = 'scatter', x = 'num_pets', y = 'num_children')
plt.show()
```

### ploting bar 

```python
df.plot(kind = 'hist', bins = [x for x in range(min(df.age.values), max(df.age.values)+1, 10 )], x = 'name', y = 'age', rwidth = 0.7)
plt.show()
```

### ploting 2 infos in one graph 

```python
ax = plt.gca()
df.plot(kind = 'line', x = 'name', y = 'num_pets', ax = ax)
df.plot(kind = 'line', x = 'name', y = 'num_children',color = 'yellow', ax = ax)
plt.show()
```

```python
df.groupby('state')['name'].nunique().plot(kind='bar')
plt.show()
```

```python
df.groupby(['state','gender']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
df.groupby(['gender','state']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

```python
from random import randint
from random import uniform
from random import choice
countries = ['BR', 'USA', 'ME', 'CL', 'TW']
df2 = pd.DataFrame({'DATE':[randint(1, 31) for x in range(31)],
                    'COUNTRY':[choice(countries) for x in range(31)],
                    'VALUE':[round(uniform(0.1, 5.0), 2) for x in range(31)]
})
df2.head()
```

```python
ax = plt.gca()
for c in countries:
    df2[df2['COUNTRY'] == c].sort_values(by=['DATE']).plot(kind = 'line', x = 'DATE', xlim= (1, 20), y = 'VALUE', label = c , ax = ax)
plt.show()
```
