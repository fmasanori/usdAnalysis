```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
dfusd = pd.read_csv('daily.csv')
dfusd.sample(10)
```

```python
dfbtc = pd.read_csv('bitcoin_price.csv')
```

```python
listC = list(set(dfusd['Country']))
print( *list(map( lambda s: s+'\n', listC )), sep ='' )
dfusd.info()
```

```python
dfusd['Date'] = pd.to_datetime(dfusd['Date'],
                              format='%Y %m %d')
dfusd.info()
```

```python
for c in listC:
    dfusd[dfusd['Country'] == c].plot(kind= 'line', x = 'Date', y= 'Exchange rate', label = c,
                                      ax = plt.gca(), figsize = (20,10), 
                                      ylim = (0, 10)).set_xlim(datetime(2014,1,1))
plt.show()
```
