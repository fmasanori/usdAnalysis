```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
dfusd = pd.read_csv('daily.csv')
dfusd.sample(10)
```

```python
dfusd.head()
```

```python
listC = list(set(dfusd['Country']))
print(listC)
dfusd.info()
```

```python
dfusd['Date'] = pd.to_datetime(dfusd['Date'],
                              format='%Y %m %d')
dfusd.info()
```

```python
for c in listC:
    dfusd[dfusd['Country'] == c].plot(kind= 'line', x = 'Date', y= 'Exchange rate', label = c,
                                      ax = plt.gca(), figsize = (20,10))
plt.show()
```

## Crise econômica de 2007-2008

```python
for c in listC:
    dfusd[dfusd['Country'] == c].plot(kind= 'line', x = 'Date', y= 'Exchange rate', label = c,
                                      ax = plt.gca(), figsize = (20,10), 
                                      ylim = (0, 5)).set_xlim(left=datetime(2006,1,1),right=datetime(2009,12,31))
plt.show()
```

## 11 de setembro

```python
for c in listC:
    dfusd[dfusd['Country'] == c].plot(kind= 'line', x = 'Date', y= 'Exchange rate', label = c,
                                      ax = plt.gca(), figsize = (20,10), 
                                      ylim = (0, 5)).set_xlim(left=datetime(1999,1,1), right=datetime(2000,12,31))
plt.show()
```

```python
listP = ['Japan', 'Norway', 'New Zealand', 'Australia', 'India']
for c in listP:
    dfusd[dfusd['Country'] == c].plot(kind= 'line', x = 'Date', y= 'Exchange rate', label = c,
                                      ax = plt.gca(), figsize = (20,10), 
                                      ylim = (0, 3))
plt.show()
```

## Tigres asiáticos

```python
listTigers = ['Taiwan', 'South Korea', 'Hong Kong', 'Singapore']
for c in listTigers:
    dfusd[dfusd['Country'] == c].plot(kind= 'line', x = 'Date', y= 'Exchange rate', label = c, title = 'Tigres Asiáticos',
                                      ax = plt.gca(), figsize = (20,10), 
                                      ylim = (600, 2000))
plt.show()
```
